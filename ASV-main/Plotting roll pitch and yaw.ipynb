```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
odd_lines=[]
even_lines= []
new=""

with open("TC_1.9_0.7ms_1.txt") as fp:
    for i, line in enumerate(fp):
        if i%2 == 0:
            even_lines.append(line.strip())
        
        if i%2 == 1:
            odd_lines.append(line)
        
            
for j in range(0,6603):
    new+=even_lines[j]+ odd_lines[j]
```

```python
len(odd_lines)
```

```python
with open("datafile.csv", "w") as file:

    file.write(new)
```

```python
df= pd.read_csv("datafile.csv")
df
```

```python
roll= np.arctan(2*(df.q0 * df.q1 + df.q2* df.q3 )/(1-2*(df.q1**2 + df.q2**2)))

%matplotlib inline
plt.plot(df.time, roll)
plt.xlabel('Time')
plt.ylabel('Roll')
```

```python
pitch= np.arcsin(2*(df.q0*df.q2- df.q3*df.q1))

%matplotlib inline
plt.plot(df.time,pitch)


plt.xlabel('Time')
plt.ylabel('Pitch')
```

```python
yaw= np.arctan(2*(df.q0 * df.q3 + df.q1* df.q2 )/(1-2*(df.q2**2 + df.q3**2)))

%matplotlib inline
plt.plot(df.time,yaw)

plt.xlabel('Time')
plt.ylabel('Yaw')
```
